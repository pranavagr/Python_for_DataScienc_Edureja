# 2. Data Manipulation:

```python
# a. Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.
# [Hint: Use sum method with appropriate axis]


#using iloc : index location to slice the dataframe for required columns
df_prn["total_benefitted"] = df_prn.iloc[:, 2:6 ].sum(axis=1) # Index 2 to 5; 6 not included
df_prn
```

```python
df_prn.columns
```

```python
# b. Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.
df_prn = df_prn.append(df_prn.sum(), ignore_index=True)
df_prn.at[35, "STATE/UT"] =  "totals"
df_prn.at[35, "YEAR"] = 2013
df_prn
```

# 3. Plotting

```python
""" a. Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. 
Which state has the maximum number of beneficiaries?
[Hint: Use bar method of pyplot] """

#Dropping the last row with totals
df_prn.drop(df_prn.tail(1).index, inplace = True)

x_state_ut = df_prn["STATE/UT"]
y_total_benefitted  = df_prn["total_benefitted"]

plt.figure(figsize=(20,10))
plt.bar(x_state_ut, y_total_benefitted)
plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.title("STATE/UT vs Total Benefitted Inmates")
plt.xlabel("STATE/UT")
plt.ylabel("Total Benefitted Inmates")
plt.show()
```

# The maximum number of beneficiaries is in the state of Andhra Pradesh.

```python
# b. Make a pie chart that depicts the ratio among different modes of benefits.
# [Hint: Use pie method of pyplot]

values = df_prn["total_benefitted"]
label = df_prn["STATE/UT"]
plt.figure(figsize=(14,10))
plt.pie(values, labels=label, autopct='%1.1f%%')
#plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.show()
```
