```python
#Defining Features and target variable
X = df_horse.drop(["outcome"], axis=1)
y = df_horse.outcome
```

```python
"""2. This dataset contains many categorical features, replace them with label encoding.
[Hint: Refer to get_dummies methods in pandas dataframe or Label encoder in scikit-learn]"""
X_num = pd.get_dummies(X)
X_num.head()
```

```python
#Imputing values
X_num = X_num.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
X_num.head()
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X_num , y, test_size=0.3, random_state=40)
```

```python
#4. Fit a decision tree classifier and observe the accuracy.
from sklearn.tree import DecisionTreeClassifier as dtc
dtc = dtc().fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"Score of DecisionTreeClassifier: {accuracy_score(y_pred, y_test)}")
```

```python
#5. Fit a random forest classifier and observe the accuracy.
from sklearn.ensemble import RandomForestClassifier as rfc
rfc = rfc(n_estimators=100).fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"Score of RandomForestClassifier: {accuracy_score(y_pred, y_test)}")
```
