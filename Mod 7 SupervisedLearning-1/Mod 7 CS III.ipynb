```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_loan = pd.read_csv("loan_borowwer_data.csv", header=0)
df_loan.shape
```

```python
#Analyzing the count of the outcome for the Purpose column
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(14,10))
sns.countplot(x=df_loan["purpose"], hue=df_loan["not.fully.paid"])
```

```python
#Coverting the categorical variable to the numericcals
df_loan_num = pd.get_dummies(df_loan, columns= ["purpose"], drop_first=True)
df_loan_num.head()
```

```python
df_loan_num.columns
```

```python
df_loan_num.info()
```

```python
#Specifiying the Feature vector
X = df_loan_num.drop(["not.fully.paid"], axis=1)
X.head()
```

```python
y = df_loan_num["not.fully.paid"]
y.head()
```

```python
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(20,20))
sns.heatmap(X.corr(), annot=True)
#plt.show()
```

```python
# No evidence of Multicollinearity
```

```python
#Splitting the data in training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
#Predicting the Data
from sklearn.tree import DecisionTreeClassifier as dtc
dtc = dtc().fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy score: {accuracy_score(y_pred, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_pred, y_test))
```

```python
#Predicting the Data
from sklearn.ensemble import RandomForestClassifier as rfc
rfc= rfc(n_estimators=100)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy score: {accuracy_score(y_pred, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_pred, y_test))
```
