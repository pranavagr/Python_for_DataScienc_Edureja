```python
"""1.Load the data from “college.csv” that has attributes collected about private and public colleges for a particular year.
We will try to predict the private/public status of the college from other attributes."""

#Loading the dataset
data = pd.read_csv("College.csv", header=0)
data.shape
```

```python
#Feature Variable (Assigning all the columns except the target variable)
X = data.drop(["Private"], axis=1)
X.head()
```

```python
#Target Variable
y = data.Private
y.head()
```

```python
"""2. Use LabelEncoder to encode the target variable in to numerical form and 
split the data such that 20% of the data is set aside for testing."""

#Encoding the tagert variable
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(y)
print(integer_encoded)
# invert first example
inverted = label_encoder.inverse_transform([np.argmax(integer_encoded[:])])
print(inverted)
```

```python
"""3. Fit a linear svm from scikit learn and observe the accuracy.
[Hint: Use Linear SVC]"""

def linearSVC(X, integer_encoded):
    
     #Splitting the data in train-test with 20% of the data kept aside for testing.
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, integer_encoded, test_size=0.2, random_state=30)
    print(f"Splitting the data with test-size as 20% and random state as 30")

    from sklearn.svm import LinearSVC
    clf_linSVC = LinearSVC()
    clf_linSVC = clf_linSVC.fit(X_train, y_train)

    #Predicting the data
    y_pred = clf_linSVC.predict(X_test)

    import sklearn.metrics as metrics
    score_SVC = metrics.accuracy_score(y_pred, y_test)
    print(f"The Accuracy score with the SVC classifier is : {score_SVC}")

    #Printing the Confusion matrix for the SVC classifier
    cls_SVC = metrics.confusion_matrix(y_pred, y_test)
    print(f"Confusion matrix with the SVC classifier is: \n{cls_SVC}")
    
    #Printing the Classification report for the SVC classifier
    cls_SVC = metrics.classification_report(y_pred, y_test)
    print(f"Classification report with the SVC classifier is: \n{cls_SVC}")
    print('*'*40)
```

```python
print("Predicting values using the SVC classifier")
linearSVC(X, integer_encoded)
```

```python
"""4. Preprocess the data using StandardScalar and fit the same model again and observe the change in accuracy.
[Hint: Refer to scikitlearn’s preprocessing methods]"""

from sklearn.preprocessing import StandardScaler
scaled = StandardScaler().fit_transform(X) 
scaled_df = pd.DataFrame(scaled, columns = X.columns)
scaled_df.head()
```

```python
print("Predicting values using the SVC classifier after StandardScalar normalization")
linearSVC(scaled_df, integer_encoded)
```

```python
"""5. Use scikit learn’s gridsearch to select the best hyperparameter for a non-linear SVM, 
identify the model with best score and its parameters.
[Hint: Refer to model_selection module of Scikit learn]"""

#Splitting the data in train-test with 20% of the data kept aside for testing.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, integer_encoded, test_size=0.2, random_state=30)
print(f"Splitting the data with test-size as 20% and random state as 30")

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

parameter_candidates = [
 # {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
  {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf', 'sigmoid']},
]

# Create a classifier object with the classifier and parameter candidates
clf_grid = GridSearchCV(estimator=SVC(), param_grid=parameter_candidates, n_jobs=-1)

# Train the classifier 
clf_grid = clf_grid.fit(X_train, y_train)
print(f"{parameter_candidates}")
```

```python
# View the best parameters for the model found using grid search
print('Best C:',clf_grid.best_estimator_.C) 
print('Best Kernel:',clf_grid.best_estimator_.kernel)
print('Best Gamma:',clf_grid.best_estimator_.gamma)
```
