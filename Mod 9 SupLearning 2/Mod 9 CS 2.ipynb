```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
"""1.Load the kinematics dataset as measured on mobile sensors from the file “run_or_walk.csv”. 
List out the columns in the dataset."""

#Loading the dataset
data = pd.read_csv("run_or_walk.csv", header=0)
data.shape
```

```python
"""2. Let the target variable ‘y’ be the activity and assign all the columns after it to ‘x’."""

#Target Variable
y = data.activity
y.head()
```

```python
#Feature Variable (Assigning all the columns after Activity)
X = data.iloc[:,5:11]
X.head()
```

```python
def naiveBayesClassifier(X, y):
    
    #Splitting the data in train-test with 30% of the data kept aside for testing.
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)
    print(f"Splitting the data with test-size as 30% and random state as 30")
    
    from sklearn.naive_bayes import GaussianNB
    clf_gaus = GaussianNB()
    clf_gaus =clf_gaus.fit(X_train, y_train)

    #Predicting the data
    y_pred = clf_gaus.predict(X_test)

    import sklearn.metrics as metrics
    score_Naive_bayes = metrics.accuracy_score(y_pred, y_test)
    print(f"The Accuracy score with the NB classifier is : {score_Naive_bayes}")

    #Printing the Confusion matrix for the NB classifier
    conf_nb = metrics.confusion_matrix(y_pred, y_test)
    print(f"Confusion matrix with the NB classifier is: \n{conf_nb}")
    
    #Printing the Classification report for the NB classifier
    cls_nb = metrics.classification_report(y_pred, y_test)
    print(f"Classification report with the NB classifier is: \n{cls_nb}")
    print('*'*40)
```

```python
print("Predicting values using the Naive Bayes algo")
naiveBayesClassifier(X, y)
```

```python
"""4.Repeat the model once using only the acceleration values as predictors and then using only the gyro values as predictors. 
Comment on the difference in accuracy between both the models."""

#Repeating the model with Accelaration values as predictors only.
#Feature Variable (Assigning all the  Accelaration values columns)
x_acln = data.iloc[:, 5:8]
x_acln.head()
```

```python
"""4.Repeat the model once using only the acceleration values as predictors and then using only the gyro values as predictors. 
Comment on the difference in accuracy between both the models."""

#Repeating the model with Gyro values as predictors only.
#Feature Variable (Assigning all the  Gyro values columns)
x_gyro = data.iloc[:, 8:11]
x_gyro.head()
```

```python
print("Predicting values using the Naive Bayes algo having Accelaration values as predictors only.")
naiveBayesClassifier(x_acln, y)
```

```python
print("Predicting values using the Naive Bayes algo having Gyro values as predictors only.")
naiveBayesClassifier(x_gyro, y)
```

```python
# Bottomline, the accuracy has decreased with gyro values as predictors. Hence, acceleration is a predominant feature.
# But both the features (Accelaration and gyro) are important to correctly estimate the target variable (activity)
```
