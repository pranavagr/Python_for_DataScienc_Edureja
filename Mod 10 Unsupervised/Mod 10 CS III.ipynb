```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
"""1.Load the file “zoo.data” and look at the info and first five rows. 
The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal."""

#loading the dataset
df_zoo = pd.read_csv("zoo.csv", header=0)
df_zoo.shape
```

```python
#Looking at the info
df_zoo.info()
```

```python
#Printing the First 5 rows
df_zoo.head()
```

```python
df_zoo.describe()
```

```python
"""2. Find out the unique number of high level class."""
uniq_cls = np.unique(df_zoo["class_type"])
uniq_cls
```

```python
y_true = df_zoo["class_type"]
```

```python
"""3. Use the 16-intermediate feature and perform an agglomerative clustering.
[ Hint: Refer to the agglomerative clustering module in scikit learn and set the number of clusters appropriately]"""

from sklearn.cluster import  AgglomerativeClustering

#Using the number of clusters as 7 in accordance to the class_type unique values
cls_agg = AgglomerativeClustering(n_clusters=7)
y_pred = cls_agg.fit_predict(df_zoo.iloc[:, 1:16])
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy score: {accuracy_score(y_pred, y_true)}")
```

```python
from sklearn.metrics import mean_squared_error
print(f"Mean_squared_error: {mean_squared_error(y_pred, y_true)}")
```
