```python
# Problem 1. Compute how much total salary cost has increased from year 2011 to 2014
sal_yearwise_totalPay = df_sal.filter(["Year", "TotalPay"]).groupby(["Year"]).sum()
sal_yearwise_totalPay
```

```python
totalpay_2011 = sal_yearwise_totalPay.query("Year == 2011")['TotalPay']
totalpay_2014 = sal_yearwise_totalPay.query("Year == 2014")['TotalPay']
salary_incr_2011_2014 = (float(totalpay_2014)  - float(totalpay_2011))
print(f"Increase in Salary from 2011 to 2014 is : {salary_incr_2011_2014}")
```

```python
# Problem 2. Which Job Title in Year 2014 has highest mean salary?
df_sal_job_mean_2014 = df_sal.query("Year == 2014").filter(["JobTitle", "Year", "TotalPay"]).groupby("JobTitle").mean()
df_sal_job_mean_2014_sorted = df_sal_job_mean_2014.sort_values(["TotalPay"], ascending=False)
df_sal_job_mean_2014_sorted
```

```python
df_sal_job_mean_2014_max = df_sal_job_mean_2014_sorted[df_sal_job_mean_2014_sorted["TotalPay"] == df_sal_job_mean_2014_sorted["TotalPay"].max()]
df_sal_job_mean_2014_max
```

```python
#print(str(df_sal_job_mean_2014_max))
print(f"The Job Title in Year 2014 has highest mean salary is: {df_sal_job_mean_2014_max.index[0]}")#
```

```python
#3. How much money could have been saved in Year 2014 by stopping OverTimePay?
totalovertime_sal_2014 = df_sal.query("Year == 2014").filter(["OvertimePay" ,"Year"]).groupby(["Year"]).sum()
totalovertime_sal_2014 = float(totalovertime_sal_2014["OvertimePay"])

print(f"Amount can be saved by stopping OvertimePay in year 2014 is: {totalovertime_sal_2014}")
```

```python
#4. Which are the top 5 common job in Year 2014 and how much do they cost SFO ?
df_sal_common_job_2014 = df_sal.query("Year == 2014").filter(["JobTitle", "TotalPayBenefits"]).groupby(["JobTitle"]).count()
df_sal_common_job_2014 = df_sal_common_job_2014.sort_values(["TotalPayBenefits"], ascending=False)
df_sal_5common_job_2014 = df_sal_common_job_2014.head(5)
print(f"Top 5 Common job in year 2014 are as follows: {df_sal_5common_job_2014}")
```

```python
top5_df_sal_job_2014 = df_sal.query("Year == 2014").filter(["JobTitle", "TotalPayBenefits"]).groupby(["JobTitle"]).sum()
print(top5_df_sal_job_2014.head(5))
```

```python
sum =0
for i in df_sal_5common_job_2014.index:
    frame = top5_df_sal_job_2014[top5_df_sal_job_2014.index==i]
    sum += frame.values.sum()
    
print(f"Total cost of the 5 top jobs in year 2014 is {sum}")    
```

```python
# 5. Who was the top earning employee across all the years?
emp1oyees_total_earnings = df_sal.filter(["EmployeeName", "TotalPayBenefits"]).groupby(["EmployeeName"]).sum()
top_earning_empl = emp1oyees_total_earnings.nlargest(5,["TotalPayBenefits"])
top_earning_empl
```

```python
print(f" {top_earning_empl.index[0]} is the top earning employee across all the years")
```
