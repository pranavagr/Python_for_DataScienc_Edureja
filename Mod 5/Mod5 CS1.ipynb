# Problem 1 
""" From the data provided on Hollywood movies
Find the highest rated movie in the “Quest” story type. """

```python
# Finding the highest RottenTomatoes rating for the Story type as Quest
highest_rate = df_hol_mvs[df_hol_mvs["Story"] == "Quest"]["RottenTomatoes"].max()
print(f"Highest Rate for the Quest Story Type: {highest_rate}")
#Finding all the movies having the Highest RottenTomatoes rating based on the Highest Rating given to the Quest Story type 
highest_rated_movies = df_hol_mvs[df_hol_mvs["RottenTomatoes"] == highest_rate]
print(highest_rated_movies)
```

```python
highest_rated_movies = highest_rated_movies.filter(["Movie", "Story"])
print(highest_rated_movies)
```

#Find the highest rated movie in the “Quest” story type.

# The highest rated movies in the “Quest” story type are: 1. The Hurt Locker  2. The Muppets

```python
#Problem 2 Find the genre in which there has been the greatest number of movie releases
genre_gp = df_hol_mvs.groupby("Genre").groups

x = genre_gp.keys()
y = genre_gp.values()

plt.figure(figsize=(12,6))
plt.bar(x, [len(value) for value in y])
plt.title("Number of Movies vs Genre")
plt.xlabel("Genre")
plt.ylabel("Number of labels")
plt.show()
```

```python
#Problem 3
# Print the names of the top five movies with the costliest budgets.
df_movies_highest_budget = df_hol_mvs[["Movie", "Budget"]].sort_values("Budget", ascending=False).head()
print(df_movies_highest_budget["Movie"])
```

```python
#Problem 4. 
"""Is there any correspondence between the critics’ evaluation of a movie and its acceptance by the public? 
Find out, by plotting the net profitability of a movie against the ratings it receives on Rotten Tomatoes."""

#Graph between Audience Score vs RottenTomatoes Score 
x_rt_score= df_hol_mvs["RottenTomatoes"]
y_aud_score = df_hol_mvs["AudienceScore"]
plt.figure(figsize=(6,6))
plt.scatter(x_rt_score, y_aud_score)
plt.title("AudienceScore vs RottenTomatoes Score")
plt.xlabel("RottenTomatoes")
plt.ylabel("AudienceScore")
plt.show()
```

```python
#Graph between Profitability vs RottenTomatoes Score 
x_rt_score= df_hol_mvs["RottenTomatoes"]
y_profitability = df_hol_mvs["Profitability"]
plt.figure(figsize=(6,6))
plt.scatter(x_rt_score, y_profitability)
plt.title("Profitability vs RottenTomatoes Score")
plt.xlabel("RottenTomatoes")
plt.ylabel("Profitability")
plt.ylim(0,2000)
plt.show()
```

# Is there any correspondence between the critics’ evaluation of a movie and its acceptance by the public?

# Yes, there is a relation between the audience score and the RottenTomatoes score. The audience score is directly proportional to the Rotten tomatoes score.

```python
#b) Print all the elements in upper case
series_data_upper = series_data.str.upper()
print(series_data_upper)
#Note: Missing vales (NaN) are automatically excluded for the string methods e.g. lower(), upper(), len()
```

```python
#c) Print the length of all the elements
series_data_length = series_data.str.len()
print(series_data_length)
#Note: Missing vales (NaN) are automatically excluded for the string methods e.g. lower(), upper(), len()
```

```python
#6.2: From the raw data below create a Pandas Series ' Atul', 'John ', ' jack ', 'Sam'
#a) Print all elements after stripping spaces from the left and right
#b) Print all the elements after removing spaces from the left only
#c) Print all the elements after removing spaces from the right only
```
