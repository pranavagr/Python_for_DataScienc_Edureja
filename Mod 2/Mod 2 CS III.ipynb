```python
def validateLength(ref_id):
    length = len(ref_id)
    return length == 12
```

```python
def check_alpha(ref_id):
    contains_upper = False
    for string in ref_id:
        if string.isalpha():
            contains_upper = True
            break
    return contains_upper
```

```python
def check_digit(ref_id):
    contains_digit = False
    for string in ref_id:
        if string.isnumeric():
            contains_digit = True
            break
    return contains_digit
```

```python
def check_spl_chrs(ref_id):
    contains_spl_chrs= False
    spcl_chrs = "$#@"
    for string in ref_id:
        if string in spcl_chrs:
            contains_spl_chrs = True
            break
        return contains_spl_chrs
```

```python
def validate_reference_id(ref_id):
    return (check_alpha(ref_id) and check_digit(ref_id) and validateLength(ref_id)) or check_spl_chrs(ref_id)
```

```python
import base64
```

```python
def encode(key, clear):
    enc = []
    for i in range(len(clear)):
        key_c = key[i % len(key)]
        enc_c = chr((ord(clear[i]) + ord(key_c)) % 256)
        enc.append(enc_c)
    return base64.urlsafe_b64encode("".join(enc).encode()).decode()
```

```python
def decode(key, enc):
    dec = []
    enc = base64.urlsafe_b64decode(enc).decode()
    for i in range(len(enc)):
        key_c = key[i % len(key)]
        dec_c = chr((256 + ord(enc[i]) - ord(key_c)) % 256)
        dec.append(dec_c)
    return "".join(dec)
```

```python
#Main Method:
def_password ="12345"

while True:
    print("Select any option:")
    print(f"1. Get Encrypted Reference Key.")
    print(f"2. Get Decrypted Reference ID.")
    print(f"3. Exit")
    user_input = int(input("Enter the preferred option (1/2/3): "))
    if user_input == 1:
        ref_id = input("Please enter the 12 digit Reference ID which should contain numbers, alphabets and special characters: ")
        
        if not validate_reference_id(ref_id):
            print("Invalid Reference Id. Please try again.")
            continue
        
        print(encode(def_password, ref_id))
        continue
    elif user_input == 2:
        ref_key = input("Please enter the Encrypted key to decrypt:  ")
        
        print(decode(def_password, ref_key))
        continue
        
    elif user_input == 3:
        print("Exit!")
        break
        
    else:
        print("Invalid Option. Please try again.")
        continue
```
